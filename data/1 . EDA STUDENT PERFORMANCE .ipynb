```python

import matplotlib.pyplot as plt
import seaborn as sns

# Recommended: use Seaborn theme for consistent look
sns.set_theme(style="whitegrid", context="talk")

# Safety checks (optional)
assert 'parental_level_of_education' in df.columns, "Column missing: parental_level_of_education"
assert 'test_preparation_course' in df.columns, "Column missing: test_preparation_course"
assert 'lunch' in df.columns, "Column missing: lunch"

# If there are NaNs, you can drop them for plotting
df_plot_1 = df[['parental_level_of_education','test_preparation_course']].dropna()
df_plot_2 = df[['parental_level_of_education','lunch']].dropna()

fig, ax = plt.subplots(1, 2, figsize=(20, 8))

# Plot 1: parental education vs test prep
sns.countplot(
    data=df_plot_1,
    x='parental_level_of_education',
    hue='test_preparation_course',
    palette='bright',
    ax=ax[0]
)
ax[0].set_title('Students vs test preparation course', color='black', fontsize=25)
ax[0].set_xlabel('Parental level of education', fontsize=14)
ax[0].set_ylabel('Count', fontsize=14)
# Rotate x labels to avoid overlap
ax[0].tick_params(axis='x', rotation=20)
# Add bar labels (after artists exist)
for container in ax[0].containers:
    ax[0].bar_label(container, color='black', fontsize=12)

# Plot 2: parental education vs lunch
sns.countplot(
    data=df_plot_2,
    x='parental_level_of_education',
    hue='lunch',
    palette='bright',
    ax=ax[1]
)
ax[1].set_title('Students vs lunch', color='black', fontsize=25)
ax[1].set_xlabel('Parental level of education', fontsize=14)
ax[1].set_ylabel('Count', fontsize=14)
ax[1].tick_params(axis='x', rotation=20)
for container in ax[1].containers:
    ax[1].bar_label(container, color='black', fontsize=12)

# Make layout fit
plt.tight_layout()

# Show the figure (critical in scripts/terminals)
plt.show()
```
